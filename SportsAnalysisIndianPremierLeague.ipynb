##### PLOTTING

```python
plt.subplot(1,2,1)
plt.hist(data2['win_by_runs'])
plt.xlabel('Wins by runs')
plt.ylabel('Frequency')
plt.subplot(1,2,2)
plt.hist(data2['win_by_wickets'],color='r')
plt.xlabel('Wins by wickets')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.violinplot(x=data2['season'], y=data2['win_by_wickets'])
plt.xlabel('Season')
plt.ylabel('Win by Wickets')
plt.title('Distribution of Win by Wickets for Each Season')
plt.xticks(rotation=45)
plt.subplot(1,2,2)
sns.violinplot(x=data2['season'], y=data2['win_by_runs'])
plt.xlabel('Season')
plt.ylabel('Win by Runs')
plt.title('Distribution of Win by Runs for Each Season')
plt.xticks(rotation=45)
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=1.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```

```python
sns.countplot(data2['city'])
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.title('Number of Matches Played in Each City')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(data1['total_runs'])
plt.xlabel('Total Runs')
plt.title('Distribution of Total Runs')
plt.show()
```

```python
dismissal_counts = data1['dismissal_kind'].value_counts()
plt.bar(dismissal_counts.index, dismissal_counts.values,color='g')
plt.xlabel('Dismissal Type')
plt.ylabel('Frequency')
plt.title('Frequency of Dismissal Types')
plt.xticks(rotation=45)
plt.show()
```

```python
average_runs_by_team = data1.groupby('batting_team')['batsman_runs'].mean()
plt.bar(average_runs_by_team.index, average_runs_by_team.values,color='c')
plt.xlabel('Team')
plt.ylabel('Average Runs')
plt.title('Average Runs Scored by Each Team')
plt.xticks(rotation=90)
plt.show()
```

```python
team_wins = data2['winner'].value_counts()
team_wins.plot(kind='bar',color='m')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Most Successful Teams')
plt.xticks(rotation=90)
plt.show()
```

```python
player_of_match = data2['player_of_match'].value_counts().head(10)
player_of_match.plot(kind='bar',color='y')
plt.xlabel('Player')
plt.ylabel('Number of Player of the Match Awards')
plt.title('Most Successful Players')
plt.xticks(rotation=90)
plt.show()
```

```python
bowler_wickets = data1['player_dismissed'].value_counts().head(10)
batsman_runs = data1.groupby('batsman')['batsman_runs'].sum().nlargest(10)
bowler_wickets.plot(kind='bar',color='k')
plt.xlabel('Bowler')
plt.ylabel('Number of Wickets')
plt.title('Top Bowlers')
plt.xticks(rotation=45)
plt.show()

batsman_runs.plot(kind='bar',color='aquamarine')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.title('Top Batsmen')
plt.xticks(rotation=45)
plt.show()
```

```python
fielding_stats = data1.groupby('fielder')['player_dismissed'].count().nlargest(10)
fielding_stats.plot(kind='bar',color='orange' )
plt.xlabel('Fielder')
plt.ylabel('Number of Catches/Run-outs/Stumpings')
plt.title('Top Fielders')
plt.xticks(rotation=45)
plt.show()
```

```python
team_strike_rate = data1.groupby('batting_team')['batsman_runs'].sum() / data1.groupby('batting_team')['ball'].count() * 100
team_strike_rate.plot(kind='bar',color='darkgreen')
plt.xlabel('Team')
plt.ylabel('Batting Strike Rate')
plt.title('Batting Strike Rate by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
toss_decision_win_percentage = data2.groupby('toss_decision')['winner'].value_counts(normalize=True) * 100
toss_decision_win_percentage.unstack().plot(kind='bar')
plt.xlabel('Toss Decision')
plt.ylabel('Win Percentage')
plt.title('Impact of Toss Decision')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
home_win_percentage = data2[data2['team1'] == data2['winner']]['team1'].value_counts() / data2['team1'].value_counts() * 100
away_win_percentage = data2[data2['team2'] == data2['winner']]['team2'].value_counts() / data2['team2'].value_counts() * 100
home_away_win_percentage = pd.concat([home_win_percentage, away_win_percentage], axis=1)
home_away_win_percentage.columns = ['Home Win Percentage', 'Away Win Percentage']
home_away_win_percentage.plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('Win Percentage')
plt.title('Home Advantage Analysis')
plt.xticks(rotation=90)
plt.show()
```
